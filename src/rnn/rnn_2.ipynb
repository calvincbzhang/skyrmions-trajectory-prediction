# Experiment 2

As we have seen in experiment 1, we have to train for many iterations in order for the network to even fit the training data well. In this experiment, we will simplify the task so that  we only need to predict a single coordinate.

The x-coordinate is not very interesting as it is basically linear (we will fit a model thorugh it just to check), but the y-coordinate is quite complex, hence we will see if our model performs well or not. We will then test the model on other particles and coordinates.

```python
import pandas as pd
import numpy as np
import pickle
import os
import matplotlib.pyplot as plt
%matplotlib inline

# import functions
import sys
sys.path.insert(0, '../')
from my_functions import *

from sklearn import preprocessing

import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN
from keras.callbacks import ModelCheckpoint
```

```python
# this function splits the data in sequences
# each sequence is n_steps long and outputs a single value, i.e. the next time frame
def split_sequences(sequences, n_steps):
    X, y = list(), list()
    for i in range(len(sequences)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the dataset
        if end_ix > len(sequences)-1:
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequences[i:end_ix, :], sequences[end_ix, :]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

## Read data

```python
data = pd.read_csv('../../data/processed/trajectories.csv')
```

```python
# select the particle and cooridinate to use
particle = 0
coordinate = 'y'

coordinate_df = data[data['particle'] == particle][coordinate].reset_index(drop=True)
coordinate_df.head()
```

```python
# scale the data
min_max_scaler = preprocessing.MinMaxScaler()
coordinates_scaled = coordinate_df.copy()
coordinates_scaled = min_max_scaler.fit_transform(np.reshape(coordinate_df.values.astype(float), (-1, 1)))
```

```python
# split into train and test sets
train_size = int(len(coordinates_scaled) * 0.67)
test_size = len(coordinates_scaled) - train_size
train, test = coordinates_scaled[0:train_size,:], coordinates_scaled[train_size:len(coordinates_scaled),:]
print(f'Train length: {len(train)}')
print(f'Test length: {len(test)}')
```

```python
# choose a number of time steps
n_steps = 1

# convert into input/output
X_train, y_train = split_sequences(train, n_steps)
X_test, y_test = split_sequences(test, n_steps)

# the dataset knows the number of features
n_features = X_train.shape[2]
```

## Define the model and train

```python
# define model
def create_model():
    model = Sequential()
    
    model.add(SimpleRNN(16, activation='relu', input_shape=(n_steps, n_features)))
    model.add(Dense(n_features))
    
    model.compile(optimizer='adam', loss='mae')
    
    return model
```

```python
# Include the epoch in the file name (uses `str.format`)
checkpoint_path = '../../models/rnn_2/cp-{epoch:04d}.ckpt'
checkpoint_dir = os.path.dirname(checkpoint_path)

# Create a callback that saves the model's weights every 5 epochs
cp_callback = ModelCheckpoint(
    filepath=checkpoint_path, 
    verbose=0, 
    save_weights_only=True,
    period=5)
```

```python
# Create a new model instance
model = create_model()

# Save the weights using the `checkpoint_path` format
model.save_weights(checkpoint_path.format(epoch=0))

# Train the model with the new callback
history = model.fit(
    X_train, 
    y_train,
    epochs=100, 
    callbacks=[cp_callback],
    validation_data=(X_test,y_test))
```

## Evaluate

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss']) 
plt.title('Model loss') 
plt.ylabel('Loss') 
plt.xlabel('Epoch') 
plt.legend(['Train', 'Test'], loc='upper left') 
plt.show()
```

```python
print('The model performance for X_train')
pred_train = evaluate(X_train, y_train, model)
print('The model performance for X_test')
pred_test = evaluate(X_test, y_test, model)
```

```python
pred_train_unscaled = pred_train.copy()
pred_test_unscaled = pred_test.copy()

pred_train_unscaled = min_max_scaler.inverse_transform(np.reshape(pred_train_unscaled.astype(float), (-1, 1)))
pred_test_unscaled = min_max_scaler.inverse_transform(np.reshape(pred_test_unscaled.astype(float), (-1, 1)))
```

```python
plt.figure(figsize=(20, 10))
plt.grid(True, axis='x')
plt.xticks(np.arange(0, coordinate_df.shape[0]+1, 50))
plt.ylim(0, max(data[coordinate])+1)

plt.axvline(x=train_size, ymin=0, ymax=1, label='training vs testing', color='g')
y = np.concatenate((pred_train_unscaled, pred_test_unscaled))
x = np.arange(y.shape[0]).reshape((y.shape[0], 1))

plt.plot(x, y, label='prediction ' + str(particle), color='tab:blue')
plt.plot(x, data[data['particle'] == particle][coordinate][n_steps:-n_steps], label='ground truth ' + str(particle), color='tab:orange')

plt.legend()
```

## Test with other particles and coordinates

```python
def test_other(particle=0, coordinate='x'):
    test_df = data[data['particle'] == particle][coordinate].reset_index(drop=True)
    test_scaled = min_max_scaler.fit_transform(np.reshape(test_df.values.astype(float), (-1, 1)))

    # split in input/output
    X, y = split_sequences(test_scaled, n_steps)
    
    y_pred = evaluate(X, y, model)

    y_pred_unsc = y_pred.copy()
    y_pred_unsc = min_max_scaler.inverse_transform(np.reshape(y_pred_unsc.astype(float), (-1, 1)))
    
    # plot
    plt.figure(figsize=(20, 10))
    plt.grid(True, axis='x')
    plt.xticks(np.arange(0, test_df.shape[0]+1, 50))
    plt.ylim(0, max(data[coordinate])+1)

    y = y_pred_unsc
    x = np.arange(y.shape[0]).reshape((y.shape[0], 1))

    plt.plot(x, y, label='prediction ' + str(particle), color='tab:blue')
    plt.plot(x, data[data['particle'] == particle][coordinate][n_steps:], label='ground truth ' + str(particle), color='tab:orange')

    plt.legend()
```

```python
test_other(3, 'y')
```
